```python
# Dataset loading

import pandas as pd
import numpy as np

user_detail = pd.read_excel('Bibit-Technical Data Analysis.xlsx', sheet_name = 'user_detail', na_values = [' ','  '])
acquisition_channel = pd.read_excel('Bibit-Technical Data Analysis.xlsx', sheet_name = 'acquisition_channel')
referral = pd.read_excel('Bibit-Technical Data Analysis.xlsx', sheet_name = 'referral')
first_transaction_amount = pd.read_excel('Bibit-Technical Data Analysis.xlsx', sheet_name = 'first_transaction_amount')
marketing_spending = pd.read_excel('Bibit-Technical Data Analysis.xlsx', sheet_name = 'marketing_spending')
```

```python
# Data exploration: User Detail

user_detail = user_detail.rename(columns = {'Install ':'Install','Start Registration ':'Start Registration'})
print('User detail:\n')
print(user_detail.info(),'\n')
print(user_detail.head(),'\n')
print('Missing data in user detail:\n')
print(user_detail.isna().sum())
```

```python
# Data exploration: Aquisition Channel

print('Aquisition channel:\n')
print(acquisition_channel.info(),'\n')
print(acquisition_channel.head(),'\n')
print('Missing data in aquisition channel:\n')
print(acquisition_channel.isna().sum())
```

```python
# Data exploration: Referral

print('Referral:\n')
print(referral.info(),'\n')
print(referral.head())
print('Missing data in referral:\n')
print(referral.isna().sum())
```

```python
# Data exploration: First transaction amount

print('First transaction amount:\n')
print(first_transaction_amount.info(),'\n')
print(first_transaction_amount.head(),'\n')
print('Missing data in first transaction amount:\n')
print(first_transaction_amount.isna().sum())
```

```python
# Data exploration: Marketing spending

print('Marketing spending:\n')
print(marketing_spending.info(),'\n')
print(marketing_spending)
```

```python
# Merging Dataset

user_comp = pd.merge(user_detail, acquisition_channel, how = 'inner', 
                     left_on = ['User ID','First Name','Last Name','Full Name'],
                    right_on = ['User ID','First Name','Last Name','Full Name'])

user_comp = pd.merge(left = user_comp, right = referral, how = 'left',
                    left_on = ['User ID','First Name','Last Name','Full Name'],
                    right_on = ['User ID','First Name','Last Name','Full Name'])

user_comp = pd.merge(left = user_comp, right = first_transaction_amount, how = 'left',
                    left_on = ['User ID','First Name','Last Name','Full Name'],
                    right_on = ['User ID','First Name','Last Name','Full Name'])

user_comp['First Time Transaction Amount'] = user_comp['First Time Transaction Amount'].fillna(0)
user_comp['Age Range'] = pd.qcut(user_comp['Age'], q = 8)
print(user_comp.shape,'\n')
print('')
print('Merged dataset information')
print(user_comp.info())
```

## From the Acquisition Cost Perspective

```python
# Average cost for acquisition channel
marketing_spending['Month'] = pd.to_datetime(marketing_spending['Month'])
#marketing_spending['Month'] = marketing_spending['Month'].apply(lambda x: x.date())

avg_cost = marketing_spending.copy()
#avg_cost = avg_cost.set_index('Month')
avg_cost_mon = avg_cost.resample('M', on = 'Month').mean().reset_index()
print(avg_cost_mon)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.dates as mdates
sns.set_style(style = 'darkgrid')
plt.style.use('seaborn')

fig, axs = plt.subplots(nrows = 1, ncols = 1)
axs.plot('Month','Cost (in IDR)', data = avg_cost_mon, lw = 2, color = 'red', marker = 'o')
axs.set_title('Average Cost for Acquisition Channel (in IDR)', fontsize = 15, weight = 'bold')
axs.set_xlabel('Date', fontsize = 12, weight = 'bold')
axs.set_ylabel('Average monthly cost', fontsize = 12, weight = 'bold')
axs.xaxis.set_major_locator(mdates.MonthLocator())
axs.xaxis.set_major_formatter(mdates.DateFormatter('%d %b %Y'))
labels, locations = plt.yticks()
plt.yticks(labels, labels)
plt.grid(axis = 'both', color = 'darkgray')
plt.savefig('Acquisition_Avg_Cost.png')
```

```python
# Detail of the average acquisition

marketing_spending['Month'] = marketing_spending['Month'].apply(lambda x:x.date())
marketing_spending = marketing_spending.groupby(['Month','Acquisiton Channel'])['Cost (in IDR)'].sum().unstack()

marketing_spending.plot(kind = 'bar')
plt.title('Cost Detail of each Acquisition Channel', fontsize = 15, weight = 'bold')
plt.xticks(rotation = 0)
labels, locations = plt.yticks()
plt.yticks(labels, labels)
plt.legend(frameon = True, facecolor = 'white')
plt.grid(axis = 'y', color = 'darkgray')
plt.xlabel('Date', fontsize = 12, weight = 'bold')
plt.ylabel('Cost (in IDR)', fontsize = 12, weight = 'bold')
plt.savefig('Acquisition_Cost_Detail.png')
```

## From the Customer Detail Perspective

```python
# Acquisition Channel to Users

acq_user = user_comp['Acquisition Channel'].value_counts()
print(acq_user)

def numb(pct, qua):
    n = pct/100 * np.sum(qua)
    return '{}'.format(n.round(decimals = 0).astype(int))

fig, axs = plt.subplots(nrows = 1, ncols=  1)
axs.pie(acq_user.values, labels = acq_user.index, autopct = lambda x:numb(x, acq_user.values),
       colors = ['Salmon','Yellow','Orange','lightblue'])
axs.set_title('Acquisition Channels', fontsize = 15, weight = 'bold')
plt.show()
plt.savefig('User_AC.png')
```

```python
# Average first transaction value (monthly)

avg_trans = user_comp[['First Time Transaction','First Time Transaction Amount']]
print(avg_trans.isna().sum())
avg_trans = avg_trans.resample('M', on = 'First Time Transaction').mean()
print(avg_trans)

fig, axs = plt.subplots(1,1)
axs.plot(avg_trans, lw = 2, marker = 'X', color = 'olivedrab')
axs.set_title('Monthly First Time Transaction Amount', fontsize = 15, weight = 'bold')
axs.set_xlabel('Month', fontsize = 12, weight = 'bold')
axs.xaxis.set_major_locator(mdates.MonthLocator())
axs.xaxis.set_major_formatter(mdates.DateFormatter('%b %Y'))
axs.set_ylabel('Average Amount (in IDR)', fontsize = 12, weight = 'bold')
labels, locations = plt.yticks()
plt.yticks(labels, labels)
plt.grid(axis = 'both', color = 'darkgray')
plt.show()
plt.savefig('Monthly_Transaction.png')
```

```python
# Registration Completion

user_reg = user_comp[['Full Name','Age','Registration Completed']]
user_reg['Age Range'] = pd.qcut(user_reg['Age'], q = 8)
user_reg.loc[pd.isna(user_reg['Registration Completed']),'Completed'] = 'No'
user_reg.loc[user_reg['Registration Completed'] >= '2021-01-01', 'Completed'] = 'Yes'
print(user_reg['Completed'].value_counts())
user_reg_plot = pd.crosstab(user_reg['Age Range'],user_reg['Completed'])

fig, axs = plt.subplots(1,1)
user_reg_plot.plot.bar(stacked = False, ax = axs, colormap = 'Spectral')
axs.set_title('Correlation Between Age and Registration Completion', fontsize = 15, weight = 'bold')
axs.set_xlabel('Age Range', fontsize = 12, weight = 'bold')
axs.set_ylabel('User Count', fontsize = 12, weight = 'bold')
plt.xticks(rotation = 0)
axs.grid(axis = 'y', color = 'darkgray')
plt.legend(frameon = True, facecolor = 'white')
plt.savefig('Age_vs_Reg.png')
```

```python
# Uninstall Rate

user_ui = user_comp[['Full Name','Age','Uninstall Time']]
user_ui['Age Range'] = pd.qcut(user_ui['Age'], q = 8)
user_ui.loc[pd.isna(user_ui['Uninstall Time']),'Uninstalled'] = 'No'
user_ui.loc[user_ui['Uninstall Time'] >= '2021-01-01', 'Uninstalled'] = 'Yes'
print(user_ui['Uninstalled'].value_counts())

user_ui_plot = pd.crosstab(user_ui['Age Range'],user_ui['Uninstalled'])

fig, axs = plt.subplots(1,1)
user_ui_plot.plot.bar(stacked = False, ax = axs, colormap = 'Spectral')
axs.set_title('Correlation Between Age and Uninstall Rate', fontsize = 15, weight = 'bold')
axs.set_xlabel('Age Range', fontsize = 12, weight = 'bold')
axs.set_ylabel('User Count', fontsize = 12, weight = 'bold')
plt.xticks(rotation = 0)
axs.grid(axis = 'y', color = 'darkgray')
plt.legend(frameon = True, facecolor = 'white')
plt.savefig('Age_vs_Uninstall.png')
```

```python
# Monthly transaction vs age

age_transaction = user_comp.groupby('Age Range')['First Time Transaction Amount'].mean()
print(age_transaction)

fig, axs = plt.subplots(1,1)
axs.plot(age_transaction, lw = 2, marker = 'o', color = 'purple')
axs.set_title('Average First Time Transaction Amount based on Age Range', fontsize = 15, weight = 'bold')
axs.set_xlabel('Age Range', fontsize = 12, weight = 'bold')
axs.set_ylabel('Average Amount (in IDR)', fontsize = 12, weight = 'bold')
labels, locations = plt.yticks()
plt.yticks(labels, labels)
pos = (0,1,2,3,4,5,6,7)
labels = ('(16.999, 19.0]','(19.0, 21.0]','(21.0, 23.0]','(23.0, 25.0]','(25.0, 27.0]','(27.0, 29.0]','(29.0, 35.0]',
         '(35.0, 65.0]')
plt.xticks(pos,labels, rotation = 45)
plt.grid(axis = 'both', color = 'darkgray')
plt.show()
plt.savefig('Monthly_Transaction_Age.png')
```

```python
# Correlation to referral

age_ref = user_comp[['Age Range','Referral Code']]
age_ref['Referral Code'] = age_ref['Referral Code'].fillna('unknown')

age_ref.loc[age_ref['Referral Code'] == 'unknown','Have Referral Code'] = 'No'
age_ref.loc[age_ref['Referral Code'] != 'unknown','Have Referral Code'] = 'Yes'

Referral_code_count = age_ref['Have Referral Code'].value_counts()

fig, axs = plt.subplots(1,2, figsize = (10,5))
sns.barplot(x = Referral_code_count.index, y = Referral_code_count.values, ax = axs[0])
axs[0].set_title('Users that Have Referral Code', weight = 'bold')
axs[0].set_xlabel('Some Users that have Referral Code',weight = 'bold')
axs[0].set_ylabel('User Count',weight = 'bold')
axs[0].grid(axis = 'y', color = 'darkgray')

sns.countplot(data = age_ref, x = 'Age Range', hue = 'Have Referral Code', ax = axs[1])
axs[1].set_title('Correlation Between Age and Referral Code',weight = 'bold')
plt.xticks(rotation = 45)
axs[1].set_xlabel('Age Range', weight = 'bold')
axs[1].set_ylabel('User Count',weight = 'bold')
axs[1].grid(axis = 'y',color = 'darkgray')
axs[1].legend(frameon = True, facecolor = 'white')

plt.savefig('Age_Referral.png')
```
