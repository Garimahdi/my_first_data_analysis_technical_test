## So... What factor that plays very vital part on happiness score calculation?

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style(style = 'darkgrid')
plt.style.use('seaborn')

# Correlation mapping

happy_2015_corr = happy_2015.corr()
fig, axs = plt.subplots(1,1)
sns.heatmap(data = happy_2015_corr, cmap = 'inferno', annot = True, ax = axs)
axs.set_title('Happiness Variable Correlation in 2015 World Happiness Report', fontsize = 15, weight = 'bold')
plt.show()
plt.savefig('happy_2015.png')
```

```python
import statsmodels.api as sm

x_2018 = sm.add_constant(happy_2018['GDP per capita'])
y_2018 = happy_2018['Score']

linear_mod_2018 = sm.OLS(endog = y_2018, exog = x_2018).fit()
print(linear_mod_2018.summary())

plt.figure()
plt.scatter(x = happy_2018['GDP per capita'], y = happy_2018['Score'], color = 'olivedrab')
m, b = np.polyfit(happy_2018['GDP per capita'], happy_2018['Score'], 1)
plt.plot(happy_2018['GDP per capita'], m*happy_2018['GDP per capita'] + b, color = 'black')
plt.title('Relation Between GDP and Happiness Score in 2018', fontsize = 15, weight = 'bold')
plt.xlabel('GDP/capita', fontsize = 12, weight = 'bold')
plt.ylabel('Happiness Score', fontsize = 12, weight = 'bold')
plt.show()
plt.savefig('LR_2018.png')
```

```python
x_2019 = sm.add_constant(happy_2019['GDP per capita'])
y_2019 = happy_2019['Score']

linear_mod_2019 = sm.OLS(endog = y_2019, exog = x_2019).fit()
print(linear_mod_2019.summary())

plt.figure()
plt.scatter(x = happy_2019['GDP per capita'], y = happy_2019['Score'])
m,b = np.polyfit(happy_2019['GDP per capita'], happy_2019['Score'], 1)
plt.plot(happy_2019['GDP per capita'], m*happy_2019['GDP per capita'] + b, color = 'black')
plt.title('Relation Between GDP and Happiness Score in 2019', fontsize = 15, weight = 'bold')
plt.xlabel('GDP/capita', fontsize = 12, weight = 'bold')
plt.ylabel('Happiness Score', fontsize = 12, weight = 'bold')
plt.show()
plt.savefig('LR_2019.png')
```
